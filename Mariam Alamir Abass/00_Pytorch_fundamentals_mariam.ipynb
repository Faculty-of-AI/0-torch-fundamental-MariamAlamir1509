```python
# Import torch
import torch
# Create random tensor
A_tensor = torch.rand(size=(7,7))
A_tensor, A_tensor.shape, A_tensor.ndim
```

```python
# Create another random tensor
B_tensor = torch.rand(size=(1,7))
A_tensor, B_tensor
# Perform matrix multiplication
output_1 = torch.matmul(A_tensor, B_tensor.T)
output_1, output_1.shape
```

```python
# Set manual seed
torch.manual_seed(0)
# Create two random tensors
A_tensor = torch.rand(size=(7,7))
B_tensor = torch.rand(size=(1,7))
A_tensor, B_tensor
# Matrix multiply tensors
output_2 = torch.matmul(A_tensor, B_tensor.T)
output_2, output_2.shape
```

```python
# Set random seed
torch.manual_seed(1234)

# Check for access to GPU
device = "cuda" if torch.cuda.is_available() else "cpu"
device

# Create two random tensors on GPU
C_tensor = torch.rand(size=(2, 3)).to(device)
D_tensor = torch.rand(size=(2, 3)).to(device)

print(C_tensor)
print(D_tensor)
```

```python
# Perform matmul on tensor_A and tensor_B
output_3 = torch.matmul(C_tensor, D_tensor.T)
output_3, output_3.shape
```

```python
# Find max
print(f"Maximum: {output_3.max()}")

# Find min
print(f"Minimum: {output_3.min()}")
```

```python
# Find arg max
print(f"Index where max value occurs: {output_3.argmax()}")

# Find arg min
print(f"Index where min value occurs: {output_3.argmin()}")
```
